# 2)Can you tell me the Distribution of Vehicles based on Year of Registration with the help of a plot

```python
years_of_registration =dfc["yearOfRegistration"]
plt.hist(years_of_registration,bins=range(1900,2023),  edgecolor='black')
plt.xlabel('Year of Registration')
plt.ylabel('Number of Vehicles')
plt.title('Distribution of Vehicles Based on Year of Registration')

plt.show()
```

```python
bins = list(range(1900,2025,5))
out = pd.cut(df.yearOfRegistration, bins=bins)
counts = pd.value_counts(out).sort_index()

print('YEAR INTERVAL\tFREQUENCY')
print(counts)
```

# 3) Create a plot based on the Variation of the price range by the vehicle type

```python
print(dfc.vehicleType.unique())
```

```python

price_by_vehicle_type = dfc.groupby('vehicleType')['price'].mean().reset_index()
price_by_vehicle_type = price_by_vehicle_type.sort_values(by='price', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(price_by_vehicle_type['vehicleType'], price_by_vehicle_type['price'], color='royalblue')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.title('Average Price by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Price')
plt.show()
```
