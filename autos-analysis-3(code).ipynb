# 2) What is the Average price of vehicle by fuel type and gearbox type.Give a plot

```python
avg_power=dfc.groupby(["vehicleType","gearbox"])["powerPS"].mean()
plt.subplots(figsize=(10,6))
avg_power.plot(kind="bar")
plt.xlabel("vehicleType and gearbox type")
plt.ylabel("average power")
plt.title("avg power of vehicletype and gearbox")
plt.show()
```

# 3) What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x='vehicleType', y='powerPS', hue='gearbox', palette='husl', data=dfc)
ax.set_title("Mean power of vehicles by type and gearbox")
ax.set_xlabel("VEHICLE TYPE")
ax.set_ylabel("MEAN POWER")
plt.show()
```

# 4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this

```python
avg_price = dfc.groupby(['brand', 'vehicleType'])['price'].mean().unstack()


plt.figure(figsize=(12, 8))
sns.heatmap(avg_price, annot=True, cmap='YlGnBu', fmt=".0f", cbar_kws={'label': 'Average Price'})
plt.title('Average Price by Brand and Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Brand')
plt.show()
```
